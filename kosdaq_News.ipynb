```python
datelist = []
for code in kosdaq10['종목코드'].tolist():        
    target_site = f'https://finance.naver.com/item/news_news.nhn?code={code}&page=1&sm=title_entity_id.basic&clusterId='        
    driver.get(target_site)
    pageCount = len(driver.find_elements_by_css_selector('.Nnavi > tbody > tr > td '))            
    print(code)
    tmplist = []
    for idx in range(1,pageCount):    
        target_site = f'https://finance.naver.com/item/news_news.nhn?code={code}&page={idx}&sm=title_entity_id.basic&clusterId='
        driver.get(target_site)        
        for option in driver.find_elements_by_css_selector('.date'):            
            #print(option.text[:10].replace('.','-'))
            tmplist.append(option.text[:10].replace('.','-'))
    datelist.append(tmplist)
```

```python
baseDF = kosdaq10List[['date']].head(106)
baseDF
```

```python
# datelist[0]
count = Counter(datelist[0])
df = pd.DataFrame.from_dict(count, orient='index').reset_index()
df = df.rename(columns={'index':'date', 0:'news'})
df
```

```python
tmp = pd.merge(baseDF, df, on='date', how='left')
tmp
```

```python
tmp = tmp.fillna(0)
tmp = tmp.astype({'news': int})
tmp
```

```python
# for문으로 list에 넣기
date10list = list()

for i in range(0, 10):
    # newslist를 순서대로 df로 만듬
    newscounter = Counter(datelist[i])
    df = pd.DataFrame.from_dict(newscounter, orient='index').reset_index()
    df = df.rename(columns={'index':'date', 0:'news'})
    # baseDF와 df를 merger 함
    tmp = pd.merge(baseDF, df, on='date', how='left')
    # 결측치 처리함
    tmp = tmp.fillna(0)
    tmp = tmp.astype({'news': int})
    # 리스트에 append 시킴
    date10list.append(tmp)
```
