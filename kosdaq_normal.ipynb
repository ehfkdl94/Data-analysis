```
path = 'kosdaq0320.csv'
pre_df = pd.read_csv( path, encoding='utf-8' )
pre_df.head(2)
```

```
mergeDF = pd.merge( recent_df, pre_df, on='종목코드')
```

```
mergeDF.head()
```

```
mergeDF.shape, mergeDF.dtypes
```

```
mergeDF['현재가'] = mergeDF['현재가'].apply( lambda x : x.replace(',', ''))
mergeDF['현재가'] = mergeDF['현재가'].astype('int')
```

```
mergeDF['저가'] = mergeDF['저가'].apply( lambda x : x.replace(',', ''))
mergeDF['저가'] = mergeDF['저가'].astype('int')
```

```
mergeDF['등락율'] = ((mergeDF['현재가'] - mergeDF['저가']) / mergeDF['저가']) * 100
```

```
mergeDF.head()
```

```
mergeDF.shape
```

```
mergeDF =mergeDF.replace([np.inf, -np.inf], np.nan).dropna(axis=0)
```

```
## 오른거20개
mergeDF20 = mergeDF.sort_values(by='등락율', ascending=False).head(20)
```

```
# 오른거 10개로 선별
mergeDF10 = mergeDF20.drop([mergeDF20.index[0],
                           mergeDF20.index[1],
                           mergeDF20.index[4],
                           mergeDF20.index[5],
                           mergeDF20.index[6],
                           mergeDF20.index[8],
                           mergeDF20.index[13],
                           mergeDF20.index[15],
                           mergeDF20.index[16],
                           mergeDF20.index[17]]
                           )
mergeDF10
```
